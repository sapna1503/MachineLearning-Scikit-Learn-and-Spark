```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pyspark.sql.functions as F 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
gpa_gre_file = 'gpa-gre.csv'
multiple_sites_file = 'multiple-sites.tsv'
```

# 1. GRE data - Use Scikitlearn to compute linear regression - Given a GRE score what GPA does it predict? 

```python
# Read from csv file
df = pd.read_csv(gpa_gre_file)
df['GRE'] = df.apply(lambda row: row['Verbal'] + row['Quant'], axis=1)

# Replace NaN with mean 
df.fillna(df.mean(), inplace=True)

# Split Data
xfit = df['GRE'].values
x = xfit.reshape(-1,1)
y = df['GPA'].values

# Split the dataset into the training set and test set
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size = 1/3, random_state = 0)

linearRegressor = LinearRegression()
linearRegressor.fit(xTrain, yTrain)
yPrediction = linearRegressor.predict(xTest)
r2_score = linearRegressor.score(x,y)

print('R2 score for regression model =',r2_score)
```

```python
plt.figure(figsize=(17,10))
plt.scatter(x, y, color = 'red')

# Training set
plt.figure(figsize=(17,5))
plt.scatter(xTrain, yTrain, color = 'red')
plt.plot(xTrain, linearRegressor.predict(xTrain), color = 'blue')
plt.title('GPA vs GRE Score (Training set)')
plt.xlabel('GRE Score')
plt.xticks(np.arange(0, 1600, 100))
plt.ylabel('GPA')
plt.show()

# Test set
plt.figure(figsize=(17,10))
plt.scatter(xTest, yTest, color = 'red')
plt.plot(xTrain, linearRegressor.predict(xTrain), color = 'blue')
plt.title('GPA vs GRE Score (Test set)')
plt.xlabel('GRE Score')
plt.xticks(np.arange(0, 1600, 100))
plt.ylabel('GPA')
plt.show()
```

# 2.1 Dwell Data - Using Scikitearn - mean, standard deviation of dwell time per site

```python
df_site = pd.read_csv(multiple_sites_file,sep='\t')

df_site_grouped = df_site.groupby(['site'], as_index=True)

print("Mean Dwell-time per site")
print(df_site_grouped.mean())

print("Standard deviation of dwell time per site")
print(df_site_grouped.std())
```

# 2.2 Using Spark - mean, standard deviation of dwell time per site

```python
spark = SparkSession.builder \
    .master("local") \
    .appName("Sample") \
    .getOrCreate()

df_spark = spark.read\
         .option("header",True)\
         .option("sep","\t")\
         .csv(multiple_sites_file)

df_mean = df_spark.groupby('site').agg(F.mean('dwell-time').alias("Mean(dwell-time)")).sort("site")

df_stddev = df_spark.groupby('site').agg(F.stddev('dwell-time').alias("Stddev(dwell-time)")).sort("site")

print("Mean Dwell-time per site")
df_mean.show()

print("Standard deviation of dwell time per site")
df_stddev.show()
```
