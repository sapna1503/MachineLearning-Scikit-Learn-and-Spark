```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
gpa_gre_file = '/Users/schandiramani/Desktop/Ass3/gpa-gre.csv'

# Read from csv file
df = pd.read_csv(gpa_gre_file)
df['GRE'] = df.apply(lambda row: row['Verbal'] + row['Quant'], axis=1)

# Replace NaN with mean 
df.fillna(df.mean(), inplace=True)

# Split Data
xfit = df['GRE'].values
x = xfit.reshape(-1,1)
y = df['GPA'].values

# Split the dataset into the training set and test set
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size = 1/3, random_state = 0)
```

```python
linearRegressor = LinearRegression()
linearRegressor.fit(xTrain, yTrain)
yPrediction = linearRegressor.predict(xTest)
r2_score = linearRegressor.score(x,y)

print('R2 score for regression model =',r2_score)

# Training set
plt.figure(figsize=(10,7))
plt.scatter(xTrain, yTrain, color = 'red')
plt.plot(xTrain, linearRegressor.predict(xTrain), color = 'blue')
plt.title('GPA vs GRE Score (Training set)')
plt.xlabel('GRE Score')
plt.xticks(np.arange(0, 1600, 100))
plt.ylabel('GPA')
plt.show()

# Test set
plt.figure(figsize=(10,7))
plt.scatter(xTest, yTest, color = 'red')
plt.plot(xTrain, linearRegressor.predict(xTrain), color = 'blue')
plt.title('GPA vs GRE Score (Test set)')
plt.xlabel('GRE Score')
plt.xticks(np.arange(0, 1600, 100))
plt.ylabel('GPA')
plt.show()

# Train Model
# lr = LinearRegression()
# regression_model= lr.fit(xfit,yfit)
# score = lr.score(xfit,yfit)
# print(score)
```
